```python
import matplotlib.pyplot as plt
```

```python
max_y = max(y_list)
```

```python
index = y_list.index(max_y)
```

```python
rel_x = x_list[index]
```

```python
print(f'Max Score: {max(y_list)} with test size: {rel_x}')
```

# ***Gradient Boosting regression***

```python
from sklearn import ensemble
```

```python
x_list=[]
y_list=[]
```

```python
params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}
```

```python
for size in range (10, 50, 1):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size/100, random_state=42)
    regressor = ensemble.GradientBoostingRegressor(**params).fit(X_train, y_train)
    #print(f'For train_size: {size/100} Score = {regressor.score(X_test, y_test)}')
    x_list.append(size/100)
    y_list.append(regressor.score(X_test, y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_list, y_list)
plt.title('Test size vs Score')
plt.xlabel('Test size')
plt.ylabel('Score')
plt.show()
```

```python
max_y = max(y_list)
```

```python
index = y_list.index(max_y)
```

```python
rel_x = x_list[index]
```

```python
print(f'Max Score: {max(y_list)} with test size: {rel_x}')
```

# ***SVR***

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
x_list=[]
y_list=[]
```

```python
for size in range (10, 50, 1):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size/100, random_state=42)
    regressor = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2)).fit(X_train, y_train)
    x_list.append(size/100)
    y_list.append(regressor.score(X_test, y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_list, y_list)
plt.title('Test size vs Score')
plt.xlabel('Test size')
plt.ylabel('Score')
plt.show()
```

```python
max_y = max(y_list)
index = y_list.index(max_y)
rel_x = x_list[index]
print(f'Max Score: {max(y_list)} with test size: {rel_x}')
```

# **Висновок:  Найкраща прогностична здатність моделі складає 0.21 у разі відсотка тестової виборки: 11%. SVR модель не є прийнятною та не може застосовуватися для класифікації. Потрібні інші рішення**
